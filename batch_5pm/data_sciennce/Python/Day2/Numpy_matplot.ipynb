```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from random import randint
fig = plt.figure(figsize=(14,10))
axes = plt.axes()
axes.set_xlim(0,100)
axes.set_ylim(0,100)
axes.set_xticks([var for var in range(-10,111,5)])
axes.set_yticks([var for var in range(-10,131,15)])
fig.set_facecolor('gray')
x = np.array([var for var in range(0,100,2) ])
y = np.array([randint(1,100) for var in range(0,100,2)])
z = np.array([ randint(1,100) for var in range(0,100,2)])
plt.scatter(x,y,c='b')
plt.scatter(x,z,c='g')
plt.plot(x,y,'b:',linewidth=2)
plt.plot(x,z,'g--',linewidth=2)
plt.xlabel('Days')
plt.ylabel('Students')
plt.legend(['Growth2017','Growth2018'],loc=4)
plt.title('Grras Solutions Pvt. Ltd, Jaipur')
plt.grid(True,color='r')
plt.show()
```

```python
#stack Plot
days = [ 1,2,3,4,5]
sleeping = [ 7,8,6,11,7]
working = [ 7,8,7,2,2]
eating = [ 2,3,4,3,2]
playing = [ 8,5,7,8,13]

plt.plot([],[],color='m',label='Sleeping',linewidth=5)
plt.plot([],[],color='c',label='Working',linewidth=5)
plt.plot([],[],color='r',label='Eating',linewidth=5)
plt.plot([],[],color='k',label='playing',linewidth=5)

plt.stackplot(days,sleeping,working,eating,playing,colors=['m','c','r','k'])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Stack Plot')
plt.legend()
plt.show()
```

```python
#multiplots
def gen_data(points):
    return np.exp(-points)*np.cos(2*np.pi*points)

p1 = np.arange(0.0,5.0,0.1)
p2 = np.arange(0.0,5.0,0.02)

plt.subplot(2,1,1)
plt.plot(p1,gen_data(p1),'bo',p2,gen_data(p2))

plt.subplot(2,1,2)
plt.plot(p2,np.cos(2*np.pi*p2))
plt.show()
```

```python
#multiplots
def gen_data(points):
    return np.exp(-points)*np.cos(2*np.pi*points)

p1 = np.arange(0.0,5.0,0.1)
p2 = np.arange(0.0,5.0,0.02)
plt.figure(figsize=(15,5))
plt.subplot(2,2,1)
plt.plot(p1,gen_data(p1),'bo',p2,gen_data(p2))

plt.subplot(2,2,2)
plt.plot(p2,np.cos(2*np.pi*p2))
plt.subplot(2,2,3)
plt.plot(p1,np.sin(2*np.pi*p1))
plt.subplot(2,2,4)
plt.plot(p1,gen_data(np.cos(2*np.pi*p1)),'g')

plt.show()
```

```python
#multiple graph in same plot
data1 = [ randint(1,50) for var in range(30)]
data2 = [ randint(1,50) for var in range(30)]
x = [ var for var in range(30) ]

fig = plt.figure(figsize=(15,5))
fig.patch.set_facecolor('gray')

graph1 = fig.add_subplot(2,2,1,facecolor='yellow')
graph1.plot(x,data1)
graph1.set_title('data1')

graph2 = fig.add_subplot(2,2,2,facecolor='lightgreen')
graph2.plot(x,data2)
graph2.set_title('data2')

graph3 = fig.add_subplot(2,1,2,facecolor='lightgray')
graph3.plot(sorted(data1),data2,'ro-')

plt.show()
```

```python
#animated graph 

import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import style
from random import randint
#style.available?
style.use('seaborn')
fig = plt.figure()
plot1 = fig.add_subplot(1,1,1)
x = [ var for var in range(30)]
data = [ randint(1,100) for var in range(30)]
def animate(i):
    global data,x
    data.append(randint(1,100))
    x.append(x[-1] + 1)
    plot1.clear()
    plot1.plot(x,data)
    
ani = animation.FuncAnimation(fig,animate,interval=1000)
plt.show()
#run this as a script to get live graph
```

```python
#more animatided

import time
import psutil
import matplotlib.pyplot as plt
%matplotlib notebook
plt.rcParams['animation.html'] = 'jshtml'

fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)
fig.show()
```

```python
i = 0
x,y = [],[]

while True : 
    x.append(i)
    y.append(psutil.cpu_percent())
    ax.plot(x,y,color='g')
    fig.canvas.draw()
    ax.set_xlim(left=max(0,i-50),right=i+50)
    time.sleep(0.1)
    i += 1
```
